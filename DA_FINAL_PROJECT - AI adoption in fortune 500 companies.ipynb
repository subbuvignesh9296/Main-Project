# **3. Treating outliers**
  * There is only one column Revenue_USD has many outliers because they can heavily affect mean, correlation, and machine learning models.
       

```python
# select numeric column for check outliers
num_cols = df.select_dtypes(include=["int64", 'float64']).columns
num_cols
```

```python
# Check Outliers using IQR for Revenue_USD

import matplotlib.pyplot as plt
import seaborn as sns

# copy dataset
df_Before_Outlier = df.copy()
df_After_Outlier = df.copy()

num_cols = ['Revenue_USD']

# Calculate Q1, Q3 and IQR

for col in num_cols:
    Q1 = df_After_Outlier[col].quantile(0.25)
    Q3 = df_After_Outlier[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Treat outliers using capping
    df_After_Outlier[col] = df_After_Outlier[col].clip(
        lower=lower_limit,
        upper=upper_limit
    )


# Beore and After treating outliers
plt.figure(figsize=(18,6))

# Beofre
plt.subplot(1,2,1)
sns.boxplot(data=df_Before_Outlier[num_cols])
plt.title("Before Treating Outliers")
plt.xticks(rotation=45)

# After
plt.subplot(1,2,2)
sns.boxplot(data=df_After_Outlier[num_cols])
plt.title("After Treating Outliers")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

 * Outlier in Revenue USD were identified using Inter Quartile Range(IQR) techniques. Since the dataset contains several Fortune 500 companies with exceptionally high revenues, these values were treated using upper-bound capping instead of deletion. This preserved all records while reducing the impact of extreme values. The post treatment boxplot shows a more balanced and analyzable revenue distribution.

```python
# Check Outliers using IQR for AI_ROI_Present Vs AI_MAturity_Score

import matplotlib.pyplot as plt
import seaborn as sns

# copy dataset
df_Before_Outlier = df.copy()
df_After_Outlier = df.copy()

num_cols = ['AI_ROI_Percent', 'AI_Maturity_Score']

# Calculate Q1, Q3 and IQR

for col in num_cols:
    Q1 = df_After_Outlier[col].quantile(0.25)
    Q3 = df_After_Outlier[col].quantile(0.75)

    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Treat outliers using capping
    df_After_Outlier[col] = df_After_Outlier[col].clip(
        lower=lower_limit,
        upper=upper_limit
    )


# Beore and After treating outliers
plt.figure(figsize=(18,6))

# Beofre
plt.subplot(1,2,1)
sns.boxplot(data=df_Before_Outlier[num_cols])
plt.title("Before Treating Outliers")
plt.xticks(rotation=45)

# After
plt.subplot(1,2,2)
sns.boxplot(data=df_After_Outlier[num_cols])
plt.title("After Treating Outliers")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

* AI_Maturity score were reduced after treating the outlires while AI_ROI present almost unchanged due to fewer outliers.

```python
# AI Efficiency index
df['AI_Efficiency_Index'] = (df["AI_ROI_Percent"] * df["AI_Maturity_Score"]) /100
df
```

# **AI Efficiency Index**

* This feature measures how efficiently companies generate ROI from AI investments.
# **Purpose**

# **Business Benefit**
* It Helps to identify companies getting better ROI with lower investment
effective AI investment strategies

```python
# Maturity Level
df['AI_Maturity_Level'] = pd.cut(df["AI_Maturity_Score"],
                                 bins=[0,40,70,100],
                                 labels=["Beginners", "Intermediate","Advanced"])
df
```

```python
# ROI categoty
df['ROI_Category'] = pd.cut(df['AI_ROI_Percent'],
                          bins=[-1,10,25,50],
                          labels=["Low ROI", "Medium ROI", "High ROI"])
df
```
